```python
import pandas as pd
import seaborn as sb
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
filename = '9TratamentoAgua.csv'
#df = pd.read_csv(filename, encoding = 'Latin1-error_bad_lines=False)
df = pd.read_csv(filename, sep=';', encoding='Latin1')
```

```python
df.columns = [c.replace(' ', '_') for c in df.columns]
```

```python
df.shape
#Respectivamente, quantidade de cadastros de tratamento de água empregado nos sistemas e soluções alternativas de abastecimento de água, e os dados registrados para cada cadastro.
```

```python
df.sample(10)
#Demonstração dos dados com cadastros selecionados aleatoriamente.
```

```python
cap = {'CAPTAÇÃO SUPERFICIAL': df.groupby('CAPTAÇÃO_SUPERFICIAL').size().sort_values(),
       'CAPTAÇÃO SUBTERRÂNEA': df.groupby('CAPTAÇÃO_SUBTERRÂNEA').size().sort_values(),
       'CAPTAÇÃO DE ÁGUA DE CHUVA': df.groupby('CAPTAÇÃO_DE_ÁGUA_DE_CHUVA').size().sort_values()}
dcap = pd.DataFrame(cap)
dcap.plot(kind='barh')
#Fontes de captação de água para tratamento no Brasil.
```

```python
et = {'PRÉ-OXIDAÇÃO' :df.groupby('ETAPA_PRÉ-OXIDAÇÃO').size().sort_values(),
     'MISTURA RÁPIDA/COAGULAÇÃO' :df.groupby('ETAPA_MISTURA_RÁPIDA/COAGULAÇÃO').size().sort_values(),
     'FLOCULAÇÃO' :df.groupby('ETAPA_FLOCULAÇÃO').size().sort_values(),
     'DECANTAÇÃO' :df.groupby('ETAPA_DECANTAÇÃO').size().sort_values(),
     'FLOTAÇÃO' : df.groupby('ETAPA_FLOTAÇÃO').size().sort_values(),
     'FLUORETAÇÃO' : df.groupby('ETAPA_FLUORETAÇÃO').size().sort_values(),
     'DESINFECÇÃO' :df.groupby('ETAPA_DESINFECÇÃO').size().sort_values(),}
det = pd.DataFrame(et)
det.plot(kind='barh')
#Etapas de tratamento de água geralmente utilizadas no Brasil.
```

```python
df.groupby('TIPO_DE_FILTRAÇÃO').size().sort_values().plot(kind='pie')
#Tipo de filtração de água geralmente utilizadas no Brasil.
```

```python
dsf = {'CLORO GÁS OU HIPOCLORITO' :df.groupby('DESINFECÇÃO_COM_CLORO__GÁS_OU_HIPOCLORITO').size().sort_values(),
     'CLORAMINA' :df.groupby('DESINFECÇÃO_COM_CLORAMINA').size().sort_values(),
     'DIÓXIDO CLORO' :df.groupby('DESINFECÇÃO_COM_DIÓXIDO_CLORO').size().sort_values(),
     'OZÔNIO' :df.groupby('DESINFECÇÃO_COM_OZÔNIO').size().sort_values(),
     'UV' : df.groupby('DESINFECÇÃO_COM_UV').size().sort_values(),}
ddsf = pd.DataFrame(dsf)
ddsf.plot(kind='bar')
ddsf.sample(2)
#Métodos de desinfecção de água geralmente utilizadas no Brasil.
```

```python
dsf = {'CARRO PIPA':df.groupby('CARRO_PIPA').size().sort_values(),
     'CHAFARIZ':df.groupby('CHAFARIZ').size().sort_values(),
     'FONTE':df.groupby('FONTE').size().sort_values(),
     'CISTERNA':df.groupby('CISTERNA').size().sort_values(),
     'CANALIZAÇÃO': df.groupby('CANALIZAÇÃO').size().sort_values(),}
ddsf = pd.DataFrame(dsf)
ddsf.plot(kind='barh')
#Meios de suprimento de água geralmente utilizados no Brasil.
```
